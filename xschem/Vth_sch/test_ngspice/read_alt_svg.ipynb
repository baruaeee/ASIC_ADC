```python
file_location = 'therm.svg'
```

```python
with open(file_location, 'r+') as file:
    lines = file.readlines()
    file.seek(0)
    file.truncate()
    pattern = r">_(.+?)_<"
    #pattern = r">_.+_<"
    #pattern2 = r"_(.+?)_"
    for i, line in enumerate(lines):
        match = re.search(pattern, line)
        if match:
            fnd = match.group(1)
            sub = str(int(fnd)-1)
            #if fnd.isdigit():
                #sub = str(int(fnd) - 1)
            lines[i]=line.replace('_'+fnd+'_','_x'+sub+'_')
            print(lines[i])
    file.writelines(lines)
    file.close()
```

```python
import re

# Sample line
#line = "This is a _*_sample_\* line"
line = "This is a >_sample_< line"

# Define the pattern
pattern = r">_(.*?)_<"

# Search for the pattern in the line
match = re.search(pattern, line)

# If a match is found, get the value of *
if match:
    value_of_star = match.group(1)
    print("Value of *:", value_of_star)
else:
    print("Pattern not found in the line.")
```
