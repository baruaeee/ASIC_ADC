```python
# initialize the Binanace client (testnet argument is not necessary in actual binance)
client = Client(api_key, api_secret, testnet=True)
```

```python
# Read crypto prices
def read_price(crypto):
    # client.get_ticker(symbol=crypto) will get a dictionary data from which I only take 'lastPrice'
    return client.get_ticker(symbol=crypto)['lastPrice']
    # return client.get_avg_price(symbol = crypto)
print(read_price('PAXGBTC'))
```

```python
def rebalance():
    # Get Asset Balance
    PAXG = float(client.get_asset_balance(asset='PAXG')['free'])
    BTC = float(client.get_asset_balance(asset='BTC')['free'])

    # Get Asset Price
    BTCUSDT = float(read_price('BTCUSDT'))
    PAXGUSDT = float(read_price('PAXGUSDT'))
    PAXGBTC = float(read_price('PAXGBTC'))

    # Calculate current ratio
    BTC_balance = BTC*BTCUSDT
    PAXG_balance = PAXG*PAXGUSDT
    Balance_sum = BTC_balance + PAXG_balance
    BTC_frac = BTC_balance/Balance_sum
    PAXG_frac = PAXG_balance/Balance_sum

    # Make an rebalnce algorithm
    BTCPAXG = BTCUSDT/PAXGUSDT
    BTCPAXG_max = 34.57
    BTCPAXG_min = 5.59
    BTC_frac_calc = 0.1 - (0.8*(BTCPAXG-BTCPAXG_max)/(BTCPAXG_max - BTCPAXG_min))
    PAXG_frac_calc = 1 - BTC_frac_calc
    BTC_adj_D =(BTC_frac_calc*Balance_sum) - BTC_balance
    PAXG_adj_D = (PAXG_frac_calc*Balance_sum) - PAXG_balance
    BTC_adj = ((BTC_frac_calc*Balance_sum) - BTC_balance)/BTCUSDT
    PAXG_adj = ((PAXG_frac_calc*Balance_sum) - PAXG_balance)/PAXGUSDT
    BTC_trade = round(abs(BTC_adj), 4)
    PAXG_trade = round(abs(PAXG_adj), 4)

    BTC_trade_d = BTC_trade*BTCUSDT
    PAXG_trade_d = PAXG_trade*PAXGUSDT

    # execute trade/convert
    if BTC_adj < 0 and BTC_trade_d > 50:
        client.create_order(symbol = 'PAXGBTC', side = 'BUY', type='MARKET', quantity=PAXG_trade)
    elif PAXG_adj < 0 and PAXG_trade_d > 50:
        client.create_order(symbol = 'PAXGBTC', side = 'SELL', type='MARKET', quantity=PAXG_trade)
```
