```python
# Read crypto prices
def read_price(crypto):
    # client.get_ticker(symbol=crypto) will get a dictionary data from which I only take 'lastPrice'
    return client.get_ticker(symbol=crypto)['lastPrice']
    # return client.get_avg_price(symbol = crypto)
print(read_price('XRPETH'))
```

```python
def rebalance():
    # Get the Asset Balance
    SOL = float(client.get_asset_balance(asset='SOL')['free'])
    XRP = float(client.get_asset_balance(asset='XRP')['free'])
    BTC = float(client.get_asset_balance(asset='BTC')['free'])
    ETH = float(client.get_asset_balance(asset='ETH')['free'])
    PAXG = float(client.get_asset_balance(asset='PAXG')['free'])

    # Get the Asset price in USD
    SOLUSDT = float(read_price('SOLUSDT'))
    XRPUSDT = float(read_price('XRPUSDT'))
    BTCUSDT = float(read_price('BTCUSDT'))
    ETHUSDT = float(read_price('ETHUSDT'))
    PAXGUSDT = float(read_price('PAXGUSDT'))

    # Get the Asset Balance in USD
    SOL_USD_Balance = SOLUSDT*SOL
    XRP_USD_Balance = XRPUSDT*XRP
    BTC_USD_Balance = BTCUSDT*BTC
    ETH_USD_Balance = ETHUSDT*ETH
    PAXG_USD_Balance = PAXGUSDT*PAXG
    # Get the total Asset Balance in USD
    Total_USD_Balance = SOL_USD_Balance + XRP_USD_Balance + BTC_USD_Balance + ETH_USD_Balance + PAXG_USD_Balance

    # Get the fractions of the Assets
    SOL_frac = SOL_USD_Balance/Total_USD_Balance
    XRP_frac = XRP_USD_Balance/Total_USD_Balance
    BTC_frac = BTC_USD_Balance/Total_USD_Balance
    ETH_frac = ETH_USD_Balance/Total_USD_Balance
    PAXG_frac = PAXG_USD_Balance/Total_USD_Balance

    # get dynamic fraction of all Assets (Base currency BTC)
    SOLBTC = float(read_price('SOLBTC'))
    ETHBTC = float(read_price('ETHBTC'))
    XRPBTC = float(read_price('XRPBTC'))
    PAXGBTC = float(read_price('PAXGBTC'))
    SOLBTC_max = 0.00372
    SOLBTC_min = 0.00006
    ETHBTC_max = 0.11902
    ETHBTC_min = 0.00826
    XRPBTC_max = 0.0000339
    XRPBTC_min = 0.0000076
    #PAXGBTC_max = 0.25278
    #PAXGBTC_min = 0.02833

    # get SOLBTC pair fraction
    if (0.1 - (0.8*(SOLBTC-SOLBTC_max)/(SOLBTC_max - SOLBTC_min))) > 0.96:
        SOL_frac_calc_WBTC = 0.96
    elif (0.1 - (0.8*(SOLBTC-SOLBTC_max)/(SOLBTC_max - SOLBTC_min))) < 0.04:
        SOL_frac_calc_WBTC = 0.04
    else:
        SOL_frac_calc_WBTC = 0.1 - (0.8*(SOLBTC-SOLBTC_max)/(SOLBTC_max - SOLBTC_min))
    BTC_frac_calc_WSOL = 1-SOL_frac_calc_WBTC
    #print(SOL_frac_calc_WBTC, BTC_frac_calc_WSOL)

    # get ETHBTC pair fraction
    if (0.1 - (0.8*(ETHBTC-ETHBTC_max)/(ETHBTC_max - ETHBTC_min))) > 0.96:
        ETH_frac_calc_WBTC = 0.96
    elif (0.1 - (0.8*(ETHBTC-ETHBTC_max)/(ETHBTC_max - ETHBTC_min))) < 0.04:
        ETH_frac_calc_WBTC = 0.04
    else:
        ETH_frac_calc_WBTC = 0.1 - (0.8*(ETHBTC-ETHBTC_max)/(ETHBTC_max - ETHBTC_min))
    BTC_frac_calc_WETH = 1-ETH_frac_calc_WBTC
    #print(ETH_frac_calc_WBTC, BTC_frac_calc_WETH)

    # get XRPBTC pair fraction
    if ((0.1 - (0.8*(XRPBTC-XRPBTC_max)/(XRPBTC_max - XRPBTC_min)))/1.3) > 0.96:
        XRP_frac_calc_WBTC = 0.96
    elif ((0.1 - (0.8*(XRPBTC-XRPBTC_max)/(XRPBTC_max - XRPBTC_min)))/1.3) < 0.04:
        XRP_frac_calc_WBTC = 0.04
    else:
        XRP_frac_calc_WBTC = (0.1 - (0.8*(XRPBTC-XRPBTC_max)/(XRPBTC_max - XRPBTC_min)))/1.3
    BTC_frac_calc_WXRP = 1-XRP_frac_calc_WBTC
    #print(XRP_frac_calc_WBTC, BTC_frac_calc_WXRP)

    # get PAXGBTC pair fraction
    '''
    if ((0.1 - (0.8*(PAXGBTC-PAXGBTC_max)/(PAXGBTC_max - PAXGBTC_min)))/4.2) > 0.96:
        PAXG_frac_calc_WBTC = 0.96
    elif ((0.1 - (0.8*(PAXGBTC-PAXGBTC_max)/(PAXGBTC_max - PAXGBTC_min)))/4.2) < 0.04:
        PAXG_frac_calc_WBTC = 0.04
    else:
        PAXG_frac_calc_WBTC = (0.1 - (0.8*(PAXGBTC-PAXGBTC_max)/(PAXGBTC_max - PAXGBTC_min)))/4.2
    BTC_frac_calc_WPAXG = 1-PAXG_frac_calc_WBTC
    '''
    #print(PAXG_frac_calc_WBTC, BTC_frac_calc_WPAXG)
    PAXG_frac_calc_WBTC = 0.2
    BTC_frac_calc_WPAXG = 0.8

    # Adjustment/Rebalance ammounts (SOLBTC pair)
    SOL_adj = ((SOL_frac_calc_WBTC*(SOL_USD_Balance+BTC_USD_Balance)) - SOL_USD_Balance)/SOLUSDT
    SOL_trade = round(abs(SOL_adj), 2)
    #print(SOL_trade)

    # Trade/convert count for loop
    trade_count = 0
    # execute trade/convert
    if SOL_adj < 0 and SOL_trade >= 0.08:
        client.create_order(symbol = 'SOLBTC', side = 'SELL', type='MARKET', quantity=SOL_trade)
        trade_count = trade_count+1
    elif SOL_adj > 0 and SOL_trade >= 0.08:
        client.create_order(symbol = 'SOLBTC', side = 'BUY', type='MARKET', quantity=SOL_trade)
        trade_count = trade_count+1

    # Adjustment/Rebalance ammounts (ETHBTC pair)
    ETH_adj = ((ETH_frac_calc_WBTC*(ETH_USD_Balance+BTC_USD_Balance)) - ETH_USD_Balance)/ETHUSDT
    ETH_trade = round(abs(ETH_adj), 4)
    #print(ETH_trade)
    # Trade/convert count for loop
    #trade_count = 0
    # execute trade/convert
    if ETH_adj < 0 and ETH_trade >= 0.0025 and trade_count == 0:
        client.create_order(symbol = 'ETHBTC', side = 'SELL', type='MARKET', quantity=ETH_trade)
        trade_count = trade_count+1
    elif ETH_adj > 0 and ETH_trade >= 0.0025 and trade_count == 0:
        client.create_order(symbol = 'ETHBTC', side = 'BUY', type='MARKET', quantity=ETH_trade)
        trade_count = trade_count+1

    # Adjustment/Rebalance ammounts (XRPBTC pair)
    XRP_adj = ((XRP_frac_calc_WBTC*(XRP_USD_Balance+BTC_USD_Balance)) - XRP_USD_Balance)/XRPUSDT
    XRP_trade = round(abs(XRP_adj))
    #print(XRP_trade)
    # Trade/convert count for loop
    #trade_count = 0
    # execute trade/convert
    if XRP_adj < 0 and XRP_trade >= 8 and trade_count == 0:
        client.create_order(symbol = 'XRPBTC', side = 'SELL', type='MARKET', quantity=int(XRP_trade))
        trade_count = trade_count+1
    elif XRP_adj > 0 and XRP_trade >= 8 and trade_count == 0:
        client.create_order(symbol = 'XRPBTC', side = 'BUY', type='MARKET', quantity=int(XRP_trade))
        trade_count = trade_count+1

    # Adjustment/Rebalance ammounts (PAXGBTC pair)
    PAXG_adj = ((PAXG_frac_calc_WBTC*(PAXG_USD_Balance+BTC_USD_Balance)) - PAXG_USD_Balance)/PAXGUSDT
    PAXG_trade = round(abs(PAXG_adj), 4)
    # execute trade/convert
    if PAXG_adj < 0 and PAXG_trade >= 0.0025 and trade_count == 0:
        client.create_order(symbol = 'PAXGBTC', side = 'SELL', type='MARKET', quantity=PAXG_trade)
        trade_count = trade_count+1
    elif PAXG_adj > 0 and PAXG_trade >= 0.0025 and trade_count == 0:
        client.create_order(symbol = 'PAXGBTC', side = 'BUY', type='MARKET', quantity=PAXG_trade)
        trade_count = trade_count+1
    return trade_count
rebalance()
```

```python
# Print status
def prnt_status():
    XRP = float(client.get_asset_balance(asset='XRP')['free'])
    ETH = float(client.get_asset_balance(asset='ETH')['free'])
    SOL = float(client.get_asset_balance(asset='SOL')['free'])
    BTC = float(client.get_asset_balance(asset='BTC')['free'])
    PAXG = float(client.get_asset_balance(asset='PAXG')['free'])
    BTCUSDT = float(read_price('BTCUSDT'))
    SOLUSDT = float(read_price('SOLUSDT'))
    ETHUSDT = float(read_price('ETHUSDT'))
    XRPUSDT = float(read_price('XRPUSDT'))
    PAXGUSDT = float(read_price('PAXGUSDT'))

    BTC_balance = BTC*BTCUSDT
    SOL_balance = SOL*SOLUSDT
    ETH_balance = ETH*ETHUSDT
    XRP_balance = XRP*XRPUSDT
    PAXG_balance = PAXG*PAXGUSDT
    Balance_sum = BTC_balance + SOL_balance + ETH_balance + XRP_balance + PAXG_balance
    BTC_frac = BTC_balance/Balance_sum
    SOL_frac = SOL_balance/Balance_sum
    ETH_frac = ETH_balance/Balance_sum
    XRP_frac = XRP_balance/Balance_sum
    PAXG_frac = PAXG_balance/Balance_sum

    print("BTC Balance:\t" + str(round(BTC, 2)) + "\t$" + str(BTC_balance), "\n" +
          "SOL Balance:\t" + str(SOL) + "\t$" + str(SOL_balance), "\n" +
          "ETH Balance:\t" + str(ETH) + "\t$" + str(ETH_balance), "\n" +
          "XRP Balance:\t" + str(XRP) + "\t$" + str(XRP_balance), "\n" +
          "PAXG Balance:\t" + str(PAXG) + "\t$" + str(PAXG_balance), "\n")
    print("USD Total:\t", str(Balance_sum), "\n")
    print("Distribution:\nBTC\t:"+str(round(BTC_frac*100, 2))+"%" "\nSOL\t:"+str(round(SOL_frac*100, 2))+"%"+
          "\nETH\t:" + str(round(ETH_frac*100, 2)) + "%" + "\nXRP\t:" + str(round(XRP_frac*100, 2)) + "%"
                                                + "\nPAXG\t:" + str(round(PAXG_frac*100, 2)) + "%\n")
prnt_status()
```

```python
while True:
    try:
        os.system('clear')
        print('BOT is running ...')
        prnt_status()
        while True:
        
            for i in range(1, 1001):
                TC = rebalance()
                if TC > 0:
                    prnt_status()
            time.sleep(5)

    except Exception as e:
        print("Execution Error!", e)
        print("Retrying ...")
```
